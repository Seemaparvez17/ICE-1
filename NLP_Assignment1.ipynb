```python
import urllib.request
import nltk
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
nltk.download('stopwords')

```

```python
# Request to the SpaceX page
response = urllib.request.urlopen('https://en.wikipedia.org/wiki/SpaceX')
html = response.read()

```

```python
# tokenizing
soup = BeautifulSoup(html,"html5lib")
text = soup.get_text(strip=True)
tokens = [t for t in text.split()]

```

```python
# Create a list of clean tokens
clean_tokens = tokens[:]

```

```python
# Obtain stop words for english text
sr = stopwords.words('english')
```

```python

for token in tokens:
    if token in stopwords.words('english'):
        clean_tokens.remove(token)

```

```python
# Get word frequency using Natural Language Toolkit (stored as a dictionary from word to frequency)
freq= nltk.FreqDist(clean_tokens)
```

```python

# Restricting the words occurred less than 5 times
for key,val in freq.items():
  if val > 5:
        print(str(key)+ ':' + str(val))

```

```python
# Use matplotlib to show a graph of the frequency (only including first 10 items
freq.plot(10,cumulative=False)

```

```python
from collections import Counter
dictionary=Counter(clean_tokens)
import matplotlib.pyplot as plt
from wordcloud import WordCloud

cloud = WordCloud(max_font_size=80).generate_from_frequencies(dictionary)
plt.figure(figsize=(10,12))
plt.imshow(cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
h= nltk.FreqDist({key:val for key,val in freq.items() if val > 60})
plt.pie(h.values(),explode=None,radius=2,labels=h.keys(),shadow=False)         
plt.show()
```
